creating a Python object (a dictionary)

Dictionary
Dictionaries are used to store data values in key:value pairs.

For the years from 2011 to 2020, the average movie durations are 103, 101, 99, 100, 100, 95, 95, 96, 93, and 90, respectively.

> Indented block



```python
# Create the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# Create a dictionary with the two lists
movie_dict = {'years': years, 'durations': durations}

# Print the dictionary
print(movie_dict)
```

```python
movie_dict
```

Creating a DataFrame from a dictionary

```python
# Create a DataFrame from the dictionary
durations_df = pd.DataFrame(movie_dict)

# Print the DataFrame
print(durations_df)
```

A visual inspection of our data

Given that the data is continuous, a line plot would be a good choice, with the dates represented along the x-axis and the average length in minutes along the y-axis.

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(years, durations)

# Create a title
plt.xlabel('Year')
plt.ylabel('Average Length (minutes)')
plt.title("Netflix Movie Durations 2011-2020")

# Show the plot
plt.show()

```

***Filtering for movies***

```python
# Calculate frequency/count of unique values in the 'type' column
type_counts = df_Netflix['type'].value_counts()

# Print the frequency/count of unique values
print("Frequency/count of unique values in the 'type' column:")
print(type_counts)
```

```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = df_Netflix[df_Netflix['type'] == 'Movie']

# Select only the columns of interest
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

# Print the first five rows of the new DataFrame
netflix_movies_col_subset[0:5]
```

Creating a scatter plot

```python
# Create a scatter plot
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])

# Add labels and title
plt.xlabel('the year of release')
plt.ylabel('the movie duration')
plt.title('Movie Duration by Year of Release')

# Show the plot
plt.show()
```

Some of these films are under an hour long! Let's filter our DataFrame for movies with a duration under 60 minutes and look at the genres. This might give us some insight into what is dragging down the average.

```python
# Filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]

# Print the first 20 rows of short_movies
short_movies[0:10]
```

```python
# Calculate frequency/count of unique values in the 'genre' column for movies with duration less than 60 minutes
genre_counts = netflix_movies_col_subset['genre'].value_counts()

# Print the frequency/count of unique values
print("Frequency/count of unique values in the 'genre' column:")
print(genre_counts)
```

```python
# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset
for lab, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == "Children":
        colors.append("red")
    elif row['genre'] == "Documentaries":
        colors.append("blue")
    elif row['genre'] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")

# Inspect the first 10 values in your list
colors[0:10]
```

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"], c = colors)

# Create a title and axis labels
plt.title("Movie duration by year of release")
plt.xlabel("Release year")
plt.ylabel("Release year")

# Show the plot
plt.show()
```
